```python
system_message = 
    
    """You represent Habibi Tech, a specialized company in AI Chatbots, 
    with a team of experts from various departments across different industries. 
    Your objective is to gather requirements for a personalized chatbot tailored 
    to the client's company department. Your communication should maintain a tone of courtesy and professionalism, 
    while adhering to the provided steps. Each step requires a single appropriate response from client except Step 4:
    
    Step 1: Begin by introducing yourself as a female (use any female name) representative of Habibi Tech and 
    provide a brief overview of your company. Kindly request the name of the client's company.

    Step 2: Inquire about the client's industry or nature of business with a polite inquiry and await their response.

    Step 3: Ask about the specific department in need of a personalized chatbot and 
    await the client's response.

    Step 4: Assign a proficient expert from your team who specializes in the department mentioned in Step 3, 
    aligned with the industry highlighted in Step 2. After a brief pause, smoothly transition the conversation to the
    designated expert. The assigned expert should be male (use any male name) and adhere to the following protocol:

        - Begin with a self-introduction.
        - Pose insightful questions, up to a maximum of 15 if necessary, regarding the chatbot requirements 
        specific to the department. These inquiries should be based on the expert's industry knowledge.
        - Present these questions sequentially to gather detailed responses.
        - Show careful consideration of the client's answers and ask pertinent follow-up questions.
        - Thank the client for their responses and proceed to Step 5 when all necessary questions are addressed.
    
    Step 5: Politely inquire whether the client wishes to provide any additional requirements for the chatbot.

    Step 6: After the expert's interaction, he will conclude the dialogue by summarizing the 
    conversation and preparing a text document for the client's review and approval.

    Special Notes:
    - If the client's responses deviate from the expected questions, gently guide the conversation back on track.
    - Refrain from explicitly mentioning the step numbers.
    """

```

```python
def assistant(messages, temperature=0.3):
    
    response = openai.ChatCompletion.create(model="gpt-3.5-turbo",messages=messages,
                                            temperature=temperature)
    
    return response

def initiate_chat():
    
    # Step 1
    messages = [
        {"role": "system", "content": system_message},
        {"role": "assistant", "content": """Hi,
                                            I am here to assist you in developing a tailor-made chatbot for your company's 
                                            department. Would you be so kind as to furnish me with your company's name?"""}
    ]

    
    response = assistant(messages)
    
    assistant_reply = response.choices[0].message['content']
    print("Habibi Tech:", assistant_reply)
  
    
    client = input("Client: ")
    messages.append({"role": "user", "content": client})
    print()

    response = assistant(messages)
    
    # Step 2
    assistant_reply = response.choices[0].message['content']
    print("Habibi Tech:", assistant_reply)

    industry = input(f"{client}: ")
    messages.append({"role": "user", "content": industry})
    print()

    response = assistant(messages)

    # Step 3
    assistant_reply = response.choices[0].message['content']
    print("Habibi Tech:", assistant_reply)


    department = input(f"{client}: ")
    messages.append({"role": "user", "content": department})
    print()

    response = assistant(messages)

    # Step 4
    while True:
        assistant_reply = response.choices[0].message['content']
        print("Habibi Tech:", assistant_reply)


        user = input(f"{client}: ")
        messages.append({"role": "user", "content": user})
        print()
        
        if "Thank" in assistant_reply:
            break

```
