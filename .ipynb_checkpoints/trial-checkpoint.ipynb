```python
def assistant(messages):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages,
    )
    return response

def initiate_chat():
    messages = [{"role": "system", "content": system_message}]
    
    while True:
        user_input = input("You: ")  # Prompt for user input
        messages.append({"role": "user", "content": user_input})
        
        response = assistant(messages)
        assistant_reply = response.choices[0].message['content']
        print("Chatbot:", assistant_reply)
        messages.append({"role": "assistant", "content": assistant_reply})
        
        if "finalize" in assistant_reply.lower():  # Check for conversation termination
            break

initiate_chat()
```

```python
def send_message(messages):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages
    )
    return response.choices[0].message['content']

def initiate_chat():
    messages = [{"role": "system", "content": system_message}]
    step = 0
    
    while step != 6:
        user_input = input("You: ")
        messages.append({"role": "user", "content": user_input})
        
        assistant_reply = send_message(messages)
        print("Assistant:", assistant_reply)
        messages.append({"role": "assistant", "content": assistant_reply})
        
        if step == 0 and "name of your company" in assistant_reply:
            step = 1
        elif step == 1 and "industry" in assistant_reply:
            step = 2
        elif step == 2 and "specific department" in assistant_reply:
            step = 3
        elif step == 3 and "male expert" in assistant_reply:
            step = 4
        elif "finalize" in assistant_reply.lower():
            step = 6
        else:
            step += 1

initiate_chat()
```
